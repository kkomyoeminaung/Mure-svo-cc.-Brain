```python
# Start Dream Mode Background Task
import threading
def start_dream():
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    loop.run_until_complete(mure.run_dream_mode())

threading.Thread(target=start_dream, daemon=True).start()
print("🌙 Dream Mode running in background...")
```
