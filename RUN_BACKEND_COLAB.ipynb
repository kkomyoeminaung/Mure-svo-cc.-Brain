```python
# 1. Install Dependencies
!pip install -q fastapi uvicorn pydantic pyngrok nest-asyncio

import os, json
from google.colab import drive

# 2. Mount Drive (Required for rules.json)
if not os.path.exists("/content/drive"):
    drive.mount("/content/drive")

# 3. Ngrok Configuration
from pyngrok import ngrok
import nest_asyncio

# Start ngrok
try:
    # Try to get token from Colab Secrets (Left sidebar Keys icon)
    from google.colab import userdata
    NGROK_TOKEN = userdata.get("NGROK_TOKEN")
except:
    NGROK_TOKEN = ""

if not NGROK_TOKEN or NGROK_TOKEN == "YOUR_NGROK_AUTH_TOKEN":
    print("💡 Pro-tip: colab keys ထဲမှာ NGROK_TOKEN ဆိုတဲ့ key နဲ့ token ကို သိမ်းထားပေးပါ။")
    NGROK_TOKEN = input("🔑 Enter your ngrok auth token (from dashboard.ngrok.com): ")

ngrok.set_auth_token(NGROK_TOKEN)
public_url = ngrok.connect(8000)
print("="*60)
print(f"⭐ PUBLIC API URL: {public_url.public_url} ⭐")
print("React Frontend ထဲက Settings Panel မှာ အထက်ပါ URL ကို ထည့်ပေးပါ။")
print("="*60)

# 4. Start Server
nest_asyncio.apply()
os.environ["MURE_BRAIN_PATH"] = "/content/drive/MyDrive/svo cc brain/rules.json"
!uvicorn fastapi_server:app --host 0.0.0.0 --port 8000
```
